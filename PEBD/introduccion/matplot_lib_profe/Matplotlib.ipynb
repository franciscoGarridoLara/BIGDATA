```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
```

Vamos a dibujar la función y = 2 * x

```python
x = [i for i in range(5)]
x
```

```python
datos = pd.DataFrame(x, columns = ['ejex'])
```

```python
datos['ejey'] = datos.ejex * 2
datos.head()
```

```python
fig, ax = plt.subplots()
ax.plot('ejex', 'ejey', data = datos)
plt.show()
```

Otra forma de usarlo es con dos listas

```python
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]
fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

Añadir títulos y texto

```python
fig, ax = plt.subplots()
ax.plot('ejex', 'ejey', data = datos,  scalex=False,  scaley=False)
ax.set_title("y = 2 * x")
ax.set_xlim(0,10)
ax.set_ylim(0,10)
ax.legend()
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.show()
```
