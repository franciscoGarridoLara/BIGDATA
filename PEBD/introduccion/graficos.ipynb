```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#Dados los datos de las flores representamos unos puntos como x la longitud de los pétalos e y el ancho.
#muestra una figura con tres gráficos como el anterior, uno por cada especie.



#Cargamos el .csv en un objetos pandas
iris = pd.read_csv("./data/iris.csv")
#Guardamos en una nueva columna como tipo category la traducción de las categorías a números.
iris["var_number"] = iris.variety.astype("category").cat.codes

#Con plt.subplots nos devuelve la figura y el axes
fig, ax = plt.subplots()

fig.suptitle("Setosa, Versicolor, Virginica (longitud y ancho de pétalos)")
#El parámetro c es para el color y el parámetro s es para el tamaño
ax.scatter("petal.length","petal.width", c="yellow", s=50, marker="D", label="Setosa", alpha=0.5, data=iris[iris.variety == "Setosa"])
ax.scatter("petal.length","petal.width", c="red", s=50, marker="o", label="Versicolor", alpha=0.5, data=iris[iris.variety == "Versicolor"])
ax.scatter("petal.length","petal.width", c="purple", s=50, marker="X", label="Virginica", alpha=0.5, data=iris[iris.variety == "Virginica"])


ax.set_xlabel("Longitud")
ax.set_ylabel("Ancho")

ax.legend()

#Pintamos por pantalla con plt.show()
plt.show()



iris.head()


```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#1. Para el fichero "statage.csv", muestra en una sola gráfica la altura mediana de niños y niñas en función de la edad. 

# Tiene que ser parecido a esto:
# http://theinformationdiet.blogspot.com/2011/11/average-height-of-american-boys-and.html
#TODO: Mejorar la forma de pasar al plot los datos, directamente un dataframe y después entre "" los atributos del mismo.l

datos = pd.read_csv("./data/statage.csv", sep=",")

datos["Agemos"] = datos.Agemos.map(lambda e: round(e/12))

agrupados_ano = datos.groupby(by=["Agemos", "Sex"], as_index=False).mean()

hombres = pd.DataFrame(agrupados_ano[agrupados_ano.Sex == 1], columns=["Agemos","M"])
mujeres = pd.DataFrame(agrupados_ano[agrupados_ano.Sex == 2], columns=["Agemos","M"])


h_x = hombres.Agemos
h_y = hombres.M

m_x = mujeres.Agemos
m_y = mujeres.M

fig, ax = plt.subplots()

fig.suptitle("Mediana de niños y niñas según la edad")
ax.plot(h_x, h_y, label="hombres")
ax.plot(m_x,m_y, label="mujeres")

ax.set_xlabel("años")
ax.set_ylabel("altura cm")
ax.legend()

plt.show()
```

```python
# 2. Para el fichero "statage.csv", muestra una figure con dos subplots, uno para niños y otro para niñas.
#En cada uno hay que poner altura mediana, percentil 25 y percentil 75 para cada edad.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

datos = pd.read_csv("./data/statage.csv", sep=",")



fig, ax = plt.subplots(nrows = 1, ncols = 2)

fig.suptitle("Altura mediana, percentil 25 y percentil 75")

ax[0].plot("Agemos", "M", data=datos[datos.Sex==1], label="Mediana")
ax[0].plot("Agemos", "P25", data=datos[datos.Sex==1], label="P25")
ax[0].plot("Agemos", "P75", data=datos[datos.Sex==1], label="P75")

ax[0].set_title("Mujeres")
ax[0].legend()


ax[1].plot("Agemos", "M", data=datos[datos.Sex==2], label="Mediana")
ax[1].plot("Agemos", "P25", data=datos[datos.Sex==2], label="P25")
ax[1].plot("Agemos", "P75", data=datos[datos.Sex==2], label="P75")

ax[1].set_title("Hombres")
ax[1].legend()

plt.show()

```

```python
# Para el fichero insurance.csv muestra una figura con cuatro subplots, uno por cada región.
# En el eje x tiene que estar la edad y el en el eje y el coste del seguro.
# Hay que diferenciar (con colores, marcadores o tamaños) entre fumadores y no fumadores y entre hombres y mujeres.
# Para hacer la comparación más sencilla, es importante que los cuatro subplots tengan la misma escala en los ejes y que haya títulos en la figura,
# los subplots y los ejes.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

datos = pd.read_csv("./data/insurance.csv", sep=",")

datos["color_fumador"] = datos.smoker.astype("category").cat.codes

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(25,12.5))

ax[0,0].scatter("age","charges", data=datos[(datos.region == "northeast") & (datos.sex == "male")], s=50, alpha=0.8, c="color_fumador", marker="X")
ax[0,0].scatter("age","charges", data=datos[(datos.region == "northeast") & (datos.sex == "female")], s=50, alpha=0.8, c="color_fumador", marker="D")

ax[0,1].scatter("age","charges", data=datos[(datos.region == "southeast") & (datos.sex == "male")], s=50, alpha=0.8, c="color_fumador", marker="X")
ax[0,1].scatter("age","charges", data=datos[(datos.region == "southeast") & (datos.sex == "female")], s=50, alpha=0.8, c="color_fumador", marker="D")

ax[1,0].scatter("age","charges", data=datos[(datos.region == "northwest") & (datos.sex == "male")], s=50, alpha=0.8, c="color_fumador", marker="X")
ax[1,0].scatter("age","charges", data=datos[(datos.region == "northwest") & (datos.sex == "female")], s=50, alpha=0.8, c="color_fumador", marker="D")

ax[1,1].scatter("age","charges", data=datos[(datos.region == "southwest") & (datos.sex == "male")], s=50, alpha=0.8, c="color_fumador", marker="X")
ax[1,1].scatter("age","charges", data=datos[(datos.region == "southwest") & (datos.sex == "female")], s=50, alpha=0.8, c="color_fumador", marker="D")

ax[0,0].set_title("northeast")
ax[0,1].set_title("southeast")
ax[1,0].set_title("northwest")
ax[1,1].set_title("southwest")
plt.show()


datos.head(100)
```
