```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#Dados los datos de las flores representamos unos puntos como x la longitud de los ptalos e y el ancho.

#Cargamos el .csv en un objetos pandas
iris = pd.read_csv("./data/iris.csv")
#Guardamos en una nueva columna como tipo category la traduccin de las categoras a nmeros.
iris["var_number"] = iris.variety.astype("category").cat.codes

#Con plt.subplots nos devuelve la figura y el axes
fig, ax = plt.subplots()

#El parmetro c es para el color y el parmetro s es para el tamao
ax.scatter("petal.length","petal.width", c="var_number", s=50, data=iris)

#Pintamos por pantalla con plt.show()
plt.show()


```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#1. Para el fichero "statage.csv", muestra en una sola grfica la altura mediana de nios y nias en funcin de la edad. 

# Tiene que ser parecido a esto:
# http://theinformationdiet.blogspot.com/2011/11/average-height-of-american-boys-and.html
#TODO: Mejorar la forma de pasar al plot los datos, directamente un dataframe y despus entre "" los atributos del mismo.l

datos = pd.read_csv("./data/statage.csv", sep=",")

datos["Agemos"] = datos.Agemos.map(lambda e: round(e/12))

agrupados_ano = datos.groupby(by=["Agemos", "Sex"], as_index=False).mean()

hombres = pd.DataFrame(agrupados_ano[agrupados_ano.Sex == 1], columns=["Agemos","M"])
mujeres = pd.DataFrame(agrupados_ano[agrupados_ano.Sex == 2], columns=["Agemos","M"])


h_x = hombres.Agemos
h_y = hombres.M

m_x = mujeres.Agemos
m_y = mujeres.M

fig, ax = plt.subplots()

fig.suptitle("Mediana de nios y nias segn la edad")
ax.plot(h_x, h_y, label="hombres")
ax.plot(m_x,m_y, label="mujeres")

ax.set_xlabel("aos")
ax.set_ylabel("altura cm")
ax.legend()

plt.show()
```
